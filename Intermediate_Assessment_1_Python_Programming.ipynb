## Challenge 1: Word Frequency Analyzer
Write a function `top_k_words(filename: str, k: int) -> list` that returns the top k frequent words.
- Ignore case and punctuation.
- Sort by frequency, then alphabetically.


```python
import re
from collections import Counter

def top_k_words(filename: str, k: int) -> list:
    # Read file
    with open(filename, "r") as f:
      text = f.read().lower()  # convert the text to lowercase to ignore the case

    # Remove punctuation and split into words
    words = re.findall(r'\b\w+\b', text) # Use regex to find all words (alphanumeric sequences)

    # Count word frequencies
    word_counts = Counter(words)

    # Sort by frequency (descending) then alphabetically (ascending)
    sorted_words = sorted(word_counts.items(), key=lambda item: (-item[1], item[0]))

    # Extract only the words from the sorted list and get the top k
    top_k = [word for word, count in sorted_words[:k]]

    return top_k

# Sample Test (create a sample file)
with open("sample.txt", "w") as f:
    f.write("Data science is evolving. Python is versatile. Learning never stops.")

print(top_k_words("sample.txt", 3))  #expected ["is", "data", "science"]
```

## Challenge 2 : Analyze daily expenses and extract insights.
User enters:

Number of transactions
Category and amount

Food 200
Travel 500
Food 300
Shopping 700

### Output
* Total expense
* Highest expense category
* Category-wise total

```python
def analyze_expenses():
    n = int(input("Enter number of transactions: "))

    expenses = {}

    for _ in range(n):
        category, amount = input().split()
        amount = float(amount)

        # Store category-wise total
        if category in expenses:
            expenses[category] += amount
        else:
            expenses[category] = amount

    # Calculate total expense
    total = sum(expenses.values())

    # Find highest expense category
    highest_category = max(expenses, key=expenses.get)

    print("Total Expense:", total)
    print("Highest Category:", highest_category)
    print("Category-wise:", expenses)

analyze_expenses()
```

### Challenge 3: Data Cleaner (Mini Pandas Logic Without Pandas)

10 ,20 ,abc ,30 ,-5 , 40

* Remove invalid values
* Remove negative values


Show:
* Clean list
* Average
* Max value

```python
def clean_data():
    n = int(input("Enter number of values: "))

    data = []

    for _ in range(n):
        value = input()

        try:
            num = float(value)

            # Ignore negative values
            if num < 0:
                continue

            data.append(num)

        except:
            # Ignore invalid values (like 'abc')
            continue

    # Calculate average safely
    if data:
        avg = sum(data) / len(data)
        max_val = max(data)
    else:
        avg = 0
        max_val = None

    print("Clean Data:", data)
    print("Average:", avg)
    print("Max:", max_val)

clean_data()
```

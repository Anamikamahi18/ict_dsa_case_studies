## 10. Finding Anomalies

Identify potential outliers in the mpg column using the Interquartile Range (IQR)
method. Specifically:

## 5. Broadcasting

Increase all horsepower values by 10% and store the updated values in a new
NumPy array. Handle missing data (if any) by replacing it with the mean of the
column before applying the increase.

```python
# Find Column Indexes
df.columns
```

```python
# Extract Horsepower Column
hp_array=df_np[:,3]
```

```python
# Handle missing values
# Convert hp_array to numeric first, coercing non-numeric values to NaN
hp_array = pd.to_numeric(hp_array, errors='coerce')
# Now np.isnan can be applied
np.isnan(hp_array).sum()
```

```python
# Replace missing values with Mean
mean_hp=np.nanmean(hp_array)
hp_array=np.where(np.isnan(hp_array),mean_hp,hp_array)
```

```python
# Apply 10% Increase (Broadcasting)
updated_hp=hp_array*1.10
```

```python
print(updated_hp[:10])
```

## 10. Conditional Aggregates

Calculate the mean mpg for cars grouped by the number of cylinders using NumPy
techniques.

```python
# Find Column Indexes
df.columns
```

```python
# Extract Columns
mpg = df_np[:, 0].astype(float)
cylinders = df_np[:, 1].astype(int)
```

```python
# Get Unique Cylinder Values
unique_cyl = np.unique(cylinders)
```

```python
# Loop and Compute Mean
for cyl in unique_cyl:
    mask = cylinders == cyl
    avg_mpg = np.mean(mpg[mask])
    print(f"Cylinders: {cyl}, Avg MPG: {avg_mpg}")
```
